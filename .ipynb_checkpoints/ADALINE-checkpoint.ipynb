# Filtro ADALINE

```python
import numpy as np
import matplotlib.pyplot as plt
import import_ipynb
from scipy.fft import fft, fftfreq
import pandas as pd
```

X2: señal deseada + ruido

X1: ruido que se desea eliminar

lr: factor de aprendizaje (learning rate)

```python
def ADALINE(X2,X1,lr): 
    n = np.arange(len(X2))
    Z = np.zeros(len(n))
    w1 = np.random.random() # se toma como valor inicial un número aleatorio entre 0 y 1
    
    for i in np.arange(len(n)):
        Y = X1[i]*w1
        Z[i] = X2[i] - Y
        delta = lr * Z[i] * X1[i]
        w1 = w1 + delta
    return Z
```

```python
# Importo los datos desde el archivo indicado
data = np.loadtxt('ecgRegistered.csv',delimiter = ',')

# Ordeno los datos importados en arrays independientes
ecgMadre = data[:,0] # Ruido que se desea eliminar
ecgRuidoso = data[:,1] # Señal ruidosa: ECG fetal + ECG materno
ecgFeto = data[:,2] # ECG fetal puro (esta señal solo se utiliza en el cálculo del error teórico)

fs = 250 # Frecuencia de muestreo
t = np.linspace(0, len(ecgMadre)/fs, len(ecgMadre)) # Vector tiempo
n = t*fs # Vector de muestras

plt.figure(figsize = (18,10))
plt.plot(n,ecgRuidoso)
plt.plot(n,ecgFeto, color = 'r')
plt.xlim(0,10000)
plt.legend(('ECG fetal + ECG materno','ECG fetal puro'), fontsize = 20)
plt.grid(True)
plt.show()
```

```python
# Learning rate = 1, muestras 0 a 15000
lr = 1
ecgFiltrado = ADALINE(ecgRuidoso,ecgMadre,lr)

plt.figure(figsize = (18,10))
plt.title('ECG Fetal Filtrado - lr = 1 - muestras 0 a 15000', size = 25)
plt.plot(n,ecgFiltrado, 'k')
plt.plot(n,ecgFeto ,'r')
plt.xlabel('n [muestras]', fontsize = 15)
plt.xlim(0,15000)
plt.ylim(-0.3,0.3)
plt.legend(('Señal filtrada', 'Señal ideal'), fontsize = 20)
plt.grid(True)
plt.show()
```

## Optimización del Learning Rate:

```python
LR=np.arange(0.00001,0.01,0.00001)
com=np.zeros((len(LR),2))
for i,lr in enumerate(LR):
    Z=ADALINE(ecgRuidoso,ecgMadre,lr)
    Error= 0.5*(ecgFeto-Z)**2
    com[i,0]=lr
    com[i,1]=np.median(Error)
    
indice_min = min(enumerate(com[:,1]), key=(lambda x: x[1]))
```

```python
import numpy as np

# Creación de un conjunto de datos para entrenamiento
trX = ecgMadre
trY = ecgRuidoso

# Definición de los ajustes y parámetros iniciales
num_steps = 100
learningRate = 0.10
criteria = 1e-8
b_0 = 1
b_1 = 1

# Proceso iterativo
for step in range(0, num_steps):
    b_0_gradient = 0
    b_1_gradient = 0
    N = float(len(trX))

    for i in range(0, len(trX)):
        b_0_gradient -= (2/N) * (trY[i] - (b_0 + b_1 * trX[i]))
        b_1_gradient -= (2/N) * (trY[i] - (b_0 + b_1 * trX[i])) * trX[i]
        
    b_0 = b_0 - (learningRate * b_0_gradient)
    b_1 = b_1 - (learningRate * b_1_gradient)

    if max(abs(learningRate * b_0_gradient), abs(learningRate * b_1_gradient)) < criteria:
        break
    
# Impresión de los resultados
print("Los valores que se obtienen son:", b_0, b_1, "en pasos", step)
```
